### Lots of Imports

```python
import os
import subprocess
from IPython.display import display, HTML, Image
import io
from operator import itemgetter
import random 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import sklearn
from sklearn import tree
from sklearn import metrics
from sklearn import model_selection
from sklearn import ensemble
from sklearn import preprocessing
from sklearn.model_selection import GridSearchCV
```

### Setup

Take only a sample of the dataset for fast testing

```python
data_sampling_rate = 0.1
```

Setup the number of folds for all grid searches (should be 5 - 10)

```python
cv_folds = 2
```

Set up a dictionary to store simple model perofrmance comparions

```python
model_test_accuracy_comparisons = dict()
model_valid_accuracy_comparisons = dict()
model_tuned_params_list = dict()
```

### Load & Partition Data

Load the dataset and explore it.

```python
file_name = r"C:\Users\Akshay\OneDrive - University College Dublin\Sem-2\ADV ML\Data\fashion-mnist_train.csv"
target_feature = "label"
num_classes = 10
classes = {0: "T-shirt/top", 1:"Trouser", 2: "Pullover", 3:"Dress", 4:"Coat", 5:"Sandal", 6:"Shirt", 7:"Sneaker", 8:"Bag", 9:"Ankle boot"}
```

```python
# Add  code here
dataset = pd.read_csv(file_name)
dataset = dataset.sample(frac=data_sampling_rate)
display(dataset.head())
```

Examine the distribution of the target levels

```python
dataset[target_feature].value_counts()
```

Display summary statistics for each feature.

```python
if(dataset.select_dtypes(include=[np.number]).shape[1] > 0):
    display(dataset.select_dtypes(include=[np.number]).describe().transpose())
if(dataset.select_dtypes(include=['object']).shape[1] > 0):
    display(dataset.select_dtypes(include=[np.object]).describe().transpose())
```

Examine presence of missing values

```python
# Check for presence of missing values
print("Missing Values")
print(dataset.isnull().sum().sort_values(ascending = False))
```

Isolate the descriptive features we are interested in

```python
X = dataset[dataset.columns[1:]]
Y = np.array(dataset[target_feature])
```

Display some of the instances in thge datset (only really useful for images)

```python
pltsize=4
row_images = 5
col_images = 5
plt.figure(figsize=(col_images*pltsize, row_images*pltsize))

for i in range(row_images * col_images):
    i_rand = random.randint(0, X.shape[0])
    plt.subplot(row_images,col_images,i+1)
    plt.axis('off')
    plt.imshow((X.iloc[i_rand]).values.astype('int').reshape(28,28), cmap='gray', vmin=0, vmax=255)
    plt.title((str(classes[Y[i_rand]])))
    # print('Training sample',i_rand,': class:',y_train[i_rand])

plt.show()
```

Split the data into a **training set**, a **vaidation set**, and a **test set**

#### First split:
#####  -80% for training + validation
#####  -20% held out as the final test set

```python
X_train_plus_valid, X_test, y_train_plus_valid, y_test = model_selection.train_test_split(
    X,
    Y,
    test_size=0.2,
    stratify=Y,
    random_state=42
)
```

#### Second split:
#####  - Split the 80% into: 60% training + 20% validation


```python
X_train, X_valid, y_train, y_valid = model_selection.train_test_split(
    X_train_plus_valid,
    y_train_plus_valid,
    test_size=0.25,
    stratify=y_train_plus_valid,
    random_state=42
)
```

```python
print("Training set size:", X_train.shape)
print("Validation set size:", X_valid.shape)
print("Test set size:", X_test.shape)
```

Normalise the data (important for some models)

```python
X_train_plus_valid = (X_train_plus_valid/255*2) - 1
X_train = (X_train/255*2) - 1
X_valid = (X_valid/255*2) - 1
X_test = (X_test/255*2) - 1
```

```python
display(X_train.shape)
display(X_train.head())
display(X_valid.shape)
display(X_valid.head())
display(X_test.shape)
display(X_test.head())
```

Normalised Image Visualisation

A random subset of training images is displayed after normalising pixel values to the range \([-1, 1]\) to confirm that the image structure is preserved.

```python
pltsize = 4
row_images = 5
col_images = 5
plt.figure(figsize=(col_images * pltsize, row_images * pltsize))

for i in range(row_images * col_images):
    i_rand = random.randint(0, X_train.shape[0] - 1)
    plt.subplot(row_images, col_images, i + 1)
    plt.axis('off')
    plt.imshow(X_train.iloc[i_rand].values.reshape(28, 28), cmap='gray', vmin=-1, vmax=1)
    plt.title(str(classes[y_train[i_rand]]))

plt.show()
```

## Train a Model

#### Random Forests

Train and evaluate a simple model random forest ensemble model.

```python
# Add code here
my_model = ensemble.RandomForestClassifier(
    n_estimators=300,
    max_features=3,
    min_samples_split=200,
    random_state=42,
    n_jobs=-1
)
my_model.fit(X_train, y_train)
```

Make and evaluate a set of predictions based on the validation data

```python
# Add code here
y_pred = my_model.predict(X_valid)
```

```python
print(metrics.classification_report(y_valid, y_pred))
sklearn.metrics.ConfusionMatrixDisplay.from_predictions(y_valid, y_pred, cmap='Blues')
```

Choose parameters using a grid search

```python
# Add code here
param_grid = [
 {
        'n_estimators': [100, 200, 300, 400, 500],
        'max_features': [2, 4, 6, 8],
        'min_samples_split': [200]
    }
]
```

```python
# Perform the search
my_tuned_model = GridSearchCV(ensemble.RandomForestClassifier(), param_grid, cv=cv_folds, verbose = 2, n_jobs = -1)
my_tuned_model.fit(X_train_plus_valid, y_train_plus_valid)

# Print details
print("Best parameters set found on development set:")
print(my_tuned_model.best_params_)
model_tuned_params_list["Tuned Random Forest"] = my_tuned_model.best_params_
print(my_tuned_model.best_score_)
```

Make and evaluate a set of predictions for the test dataset.

```python
# Add code here
y_pred = my_tuned_model.predict(X_test)
```

```python
print(metrics.classification_report(y_test, y_pred))
sklearn.metrics.ConfusionMatrixDisplay.from_predictions(y_test, y_pred, cmap = 'Blues')
```

```python
# Store accuracy results
model_test_accuracy_comparisons["Random Forest (Baseline)"] = metrics.accuracy_score(y_valid, my_model.predict(X_valid))
model_test_accuracy_comparisons["Random Forest (Tuned)"] = metrics.accuracy_score(y_test, y_pred)

# Create comparison table
accuracy_df = pd.DataFrame.from_dict(
    model_test_accuracy_comparisons,
    orient='index',
    columns=['Accuracy']
)

display(accuracy_df)
```
